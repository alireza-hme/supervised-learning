```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression


plt.style.use("default")
np.set_printoptions(precision=2)

csv_data = np.genfromtxt("../data/house_prices.csv", delimiter=",", skip_header=1)
house_data = csv_data.tolist()
X_train = np.array([house[:-1] for house in house_data])
Y_train = np.array([price[-1] for price in house_data])

model = LinearRegression()
model.fit(X_train, Y_train)

# house_features = np.array([[850, 2, 1, 36]])
# price_predictions = model.predict(house_features)
price_predictions = model.predict(X_train)

# Visualization
plt.figure(figsize=(10, 6))

# Assuming X_train[:, 0] is a relevant feature for visualization
plt.scatter(X_train[:, 0], Y_train, color='blue', label='Actual Prices')
plt.plot(X_train[:, 0], price_predictions, color='red', linewidth=2, label='Fitted Line')

plt.xlabel('House Size')
plt.ylabel('Price')
plt.title("price prediction based on size")
plt.legend()

plt.show()


print(price_predictions)
```
